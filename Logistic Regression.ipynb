```python
#                     Implementation of Logistic Regression from Scratch Using Gradient Descent 
# We can't use Closed Form Solution to minimize cost function of Logistic regression . 
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state = 41, hypercube=False, class_sep = 20)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c = y, cmap='winter', s=100)
plt.show()
```

```python
from IPython.display import Image
Image(filename='images/weight update in logistic regression, batch gradient descent.png')
# in figure below:
# n is total no of training examples
# y_i is actual label
# y_i_hat predicted probability of training example calculated using sigmoid function 
# X_i is feature vector 
```

```python
#                             Function to Update Parameters of Logistic Regression using batch gradient descent
def sigmoid(z):
    return 1/(1 + np.exp(-z))

def gd(X, y):
    
    X = np.insert(X, 0, 1, axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5
    
    for i in range(2000):
        y_hat = sigmoid(np.dot(X, weights))
        weights = weights + lr*(np.dot((y - y_hat), X) / X.shape[0])
        
    return weights[1:], weights[0]
```

```python
coef_, intercept_ = gd(X, y)

m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
print('m and b are', m, b)

x_input1 = np.linspace(-3, 3, 100)
y_input1 = m * x_input1 + b

plt.figure(figsize=(10,6))
plt.plot(x_input1, y_input1, color='red', linewidth=3)
plt.plot(x_input1, y_input1, color='black', linewidth=3)
plt.scatter(X[:, 0], X[:,1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
```

```python
# red and black line are almost overlap 
```

```python
# Now lets use scikit-learn for logistic regression 
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(penalty=None, solver='sag')
lr.fit(X, y)

# Calculate slope and intercept for decision boundary
m1 = -(lr.coef_[0][0] / lr.coef_[0][1])
b1 = -(lr.intercept_ / lr.coef_[0][1])

# Printing the slope and intercept
print(f"Slope (m1): {m1}")
print(f"Intercept (b1): {b1}")
```
